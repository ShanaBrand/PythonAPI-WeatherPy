```python
import openweathermapy.core as owm
from citipy import citipy
import json
import requests
import pandas as pd
from config import api_key
import numpy as np
import random as random
import urllib
import matplotlib.pyplot as plt
```

```python
weatherTbl = pd.DataFrame()
#weatherTbl["City Name"] = city_name
```

```python
latitude = np.random.uniform(-90,90,800)
longitude = np.random.uniform(-200,200,800)
coordinates = zip(latitude,longitude)
```

```python
city_name = []

#Variable to store API extractions based on key values
extName=[] 
extTemp=[]
extLat =[]
extLon = []
extHum =[]
extCloud =[]
extWspeed=[]
extCountry=[]
```

```python
for lat,lon in coordinates:
    city = citipy.nearest_city(lat, lon)
    if city not in city_name:
        city_name.append(city.city_name)
numCities = len(city_name)
numCities
```

```python
settings = {"units": "metric", "appid": api_key}
```

```python
#Establish Weather Table data frame that will store the extract API variable calls
weatherTbl = pd.DataFrame()
weatherTbl
```

```python
#Testing API call and response with the first city in the list of cities
#test = [owm.get_current(city_name[0],**settings)]
#test
```

```python
#For each city in the list of cities created by Citipy
#Perform an API Call to OpenweatherMap
#Extract the variables API Keys for city Name, Temperature,Latitude,Longitude,Humidity, Cloudiness,Windspeed and Country
#While looping through the Append each variables obtained from the API call to the respective list
for city in city_name:
    try:
        city_Weather=[owm.get_current(city, **settings)]
        for response in city_Weather:
            extName.append(response('name'))
            extTemp.append(response('main.temp'))
            extLat.append(response('coord.lat'))
            extLon.append(response('coord.lon'))
            #extCoor.append(zip(response('coord.lat'),response("coord.lon")))
            extHum.append(response('main.humidity'))
            extCloud.append(response('clouds.all'))
            extWspeed.append(response('wind.speed'))
            extCountry.append(response('sys.country'))
            print(owm.BASE_URL)
    except urllib.error.HTTPError as err:
        pass
```

```python
#Assign and store variables obtained from the API call and response to the Weather Table dataframe
weatherTbl["CityName"] = extName
weatherTbl["Temperature"] = extTemp
weatherTbl["Latitude"] = extLat
weatherTbl["Longitude"] = extLon

weatherTbl["Humidity"] = extHum
weatherTbl["Cloudiness"] = extCloud
weatherTbl["WindSpeed"] = extWspeed
weatherTbl["Country"] = extCountry

#check and count for duplicates based on on Row values
weatherTbl.duplicated().sum()
#weatherTbl.loc[weatherTbl.duplicated(),:]

#dropDuplicate
#weatherTbl.drop_duplicates().shape
#weatherTbl.head()

weatherTbl.to_csv("./Weather_Table.csv", index =False,header=True)
```
